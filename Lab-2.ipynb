## 1: Briefly Describe the Dataset, what information does it contain

This dataset is based on avocado prices and sales volume in different geographies within the United States on specific dates from 2015-2020. Some of the types of geographies included are state, region, metro, country, and city. It specifically contains information about Hass avocados of 3 different varieties: 4046 = small Hass (Total number of avocados with PLU 4046 sold), 4225 = large hass, 4770 = extra large hass.

The columns included are:

date: date of observation

average_price: average price of single avocado on that day of that type

total_volume: total number of avocados sold of any variety

4046: total number of avocados with PLU 4046(small Hass) sold

4225: total number of large Hass avocados sold

4770: total number of extra large Hass avocados sold

total_bags: number of bags sold

small_bags: number of small bags of avocados sold

large_bags: number of large bags of avocados sold

xlarge_bags: number of xlarge bags of avocados sold

type: type of avocado - either conventional or organic

year: year of observation

## 7. In which California regions is the price of organic versus conventional avocados most different? Support your answer with a few summary statistics AND a visualization.

```python
#average price by each california region and type
avocado_type = avocado_cali[["average_price", "type", "geography"]].groupby(['geography', 'type']).mean()
avocado_type = avocado_type.reset_index()
avocado_type
```

```python
#pivot from long to wide, type now become columns and values are avergae price for each type by geography
wide_avocado = avocado_type.pivot(index = "geography", columns = "type", values = "average_price")
wide_avocado = wide_avocado.reset_index()
wide_avocado
```

```python
#find difference in average price
wide_avocado["price_diff"] = wide_avocado["organic"] - wide_avocado["conventional"]

#find which California region had largest price difference between organic and conventional
wide_avocado = wide_avocado.sort_values(by ="price_diff", ascending = False)
wide_avocado
```

```python
(ggplot(wide_avocado, aes(x='geography', y='price_diff', fill='geography'))
+geom_bar(stat='identity')
+labs(x="California Region")
+labs(y="Avg Price Difference")
+labs(fill = "California Region")
+ggtitle("Avg Price Difference between Organic/Conventional Avocados"))
```

Based on difference in average price, San Francisco has the biggest average price difference between organic and conventional avocados in the California regions.

```python
#range = min-max
avocado_min = avocado_cali[["average_price", "type", "geography"]].groupby(['geography', 'type']).min()
avocado_min = avocado_min.reset_index()
avocado_max = avocado_cali[["average_price", "type", "geography"]].groupby(['geography', 'type']).max()
avocado_max = avocado_max.reset_index()

#find the range
avocado_range = avocado_max.merge(avocado_min, on=["geography", "type"])
avocado_range["range"] = avocado_range['average_price_x'] - avocado_range['average_price_y']
avocado_range


#make it wide
wide_avocado_range = avocado_range.pivot(index = "geography", columns = "type", values = "range")
wide_avocado_range = wide_avocado_range.reset_index()
wide_avocado_range

#differences in range
wide_avocado_range["diff_range"] = wide_avocado_range['organic'] - wide_avocado_range['conventional']
wide_avocado_range = wide_avocado_range.sort_values(by="diff_range", ascending = False)
wide_avocado_range
```

```python
(ggplot(wide_avocado_range, aes(x='geography', y='diff_range', fill='geography'))
+geom_bar(stat='identity')
+labs(x="California Region")
+labs(y="Range of Price Difference")
+labs(fill = "California Region")
+ggtitle("Range Price Difference between Organic/Conventional Avocado"))
```

## 8. The following plot shows, for all four California regions, the proportion of the average Hass avocado sales that are small, large, or extra large; conventional vs. organic. Recreate the plot; you do not have to replicate the exact finishing touches - e.g., color, theme - but your plot should resemble the content of this plot.

```python
#filter to only needed columns
avocado_cali_filtered = avocado_cali[["date", "geography", "type","Small_Hass_Sold", "Large_Hass_Sold", "Extra_Large_Hass_Sold"]]
avocado_cali_filtered
```

```python
#create dataset for organic avocados
avocado_cali_organic = avocado_cali_filtered[avocado_cali_filtered['type'] == 'organic']
avocado_cali_organic
```

```python
#create dataset for conventional avocados
avocado_cali_conventional = avocado_cali_filtered[avocado_cali_filtered['type'] == 'conventional']
avocado_cali_conventional
```

```python
#for organic avocados find the average of each size of Hass avocado sold by California region
avocado_cali_organic_gb = avocado_cali_organic[["geography", "Small_Hass_Sold", "Large_Hass_Sold", "Extra_Large_Hass_Sold"]].groupby(["geography"]).mean()
avocado_cali_organic_gb = avocado_cali_organic_gb.reset_index()

#total Hass avocado sold by region
avocado_cali_organic_gb["total"] = avocado_cali_organic_gb['Small_Hass_Sold'] + avocado_cali_organic_gb['Large_Hass_Sold'] + avocado_cali_organic_gb['Extra_Large_Hass_Sold']

#proportion of small Hass sold out of the total sold by region
avocado_cali_organic_gb["Small"] = (avocado_cali_organic_gb['Small_Hass_Sold']/avocado_cali_organic_gb['total'])*100

#proportion of large Hass sold out of the total sold by region
avocado_cali_organic_gb["Large"] = (avocado_cali_organic_gb['Large_Hass_Sold']/avocado_cali_organic_gb['total'])*100

#proportion of extra large Hass sold out of the total sold by region
avocado_cali_organic_gb["X-Large"] = (avocado_cali_organic_gb['Extra_Large_Hass_Sold']/avocado_cali_organic_gb['total'])*100

avocado_cali_organic_gb

```

```python
avocado_cali_organic_filtered = avocado_cali_organic_gb[["geography", "Small", "Large", "X-Large"]]

#add column so when merge, it is labeled organic
avocado_cali_organic_filtered["type"] = "organic"

#wide to long with a new column size for small/medium/large and values are the %
long_avocado_cali_organic = avocado_cali_organic_filtered.melt(id_vars = ['geography', 'type'], var_name = "Size", value_name = "%")
long_avocado_cali_organic
```

```python
#for conventional avocados find the average of each size of Hass avocado sold by California region
avocado_cali_conventional_gb = avocado_cali_conventional[["geography", "Small_Hass_Sold", "Large_Hass_Sold", "Extra_Large_Hass_Sold"]].groupby(["geography"]).mean()
avocado_cali_conventional_gb = avocado_cali_conventional_gb.reset_index()

#find total Hass avocado sold by California region
avocado_cali_conventional_gb["total"] = avocado_cali_conventional_gb['Small_Hass_Sold'] + avocado_cali_conventional_gb['Large_Hass_Sold'] + avocado_cali_conventional_gb['Extra_Large_Hass_Sold']

#proportion of small 
avocado_cali_conventional_gb["Small"] = (avocado_cali_conventional_gb['Small_Hass_Sold']/avocado_cali_conventional_gb['total'])*100

#proportion of large 
avocado_cali_conventional_gb["Large"] = (avocado_cali_conventional_gb['Large_Hass_Sold']/avocado_cali_conventional_gb['total'])*100

#proportion of extra large
avocado_cali_conventional_gb["X-Large"] = (avocado_cali_conventional_gb['Extra_Large_Hass_Sold']/avocado_cali_conventional_gb['total'])*100

avocado_cali_conventional_gb
```

```python
#take out columns we don't need
avocado_cali_conventional_filtered = avocado_cali_conventional_gb[["geography", "Small", "Large", "X-Large"]]

#add a column so when we merge, we know these are conventional avocados
avocado_cali_conventional_filtered["type"] = "conventional"

#wide to long creating a column size for the three % of small/medium/large, values are the % of total by region
long_avocado_cali_conventional= avocado_cali_conventional_filtered.melt(id_vars = ['geography', 'type'], var_name = "Size", value_name = "%")
long_avocado_cali_conventional
```

```python
#merge the two datasets together
avocado_pct = pd.concat([long_avocado_cali_organic, long_avocado_cali_conventional])
avocado_pct
```

```python
color_dict = {"X-Large":'#7570b3',
              "Large":'#1c9e78',
             "Small":'#d95f01'}

avocado_pct['Size'] = pd.Categorical(avocado_pct['Size'], categories=["X-Large", "Large", "Small"])


(ggplot(avocado_pct, aes(x='geography', y='%', fill = "Size"))
+geom_bar(stat = 'identity', position = 'stack')
+facet_wrap('type')
+labs(x="Region of California")
+labs(y="Proportion")
+labs(fill = "size")
+scale_y_continuous(breaks = [0,25,50,75,100], labels = ["0%", "25%", "50%", "75%", "100%"])
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+theme(figure_size=(10,6))
+ggtitle("Proportion of Average Hass Avocado Sales by Size"))
```
