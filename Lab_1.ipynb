```python
(ggplot(df1,aes(x="income", y="life_exp", fill = "four_regions"))
+geom_point(aes(size='population'))
+theme_classic()
+labs(y = "Life Expectancy", x="Income")
+scale_x_continous(breaks)
+lims(y=(0,90), x=(0,64000)))

```
