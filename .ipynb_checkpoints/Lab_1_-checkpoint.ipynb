# Task 2

## Part 1: Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

The x variable is exports as a % of GDP and y variable is imports as a % of GDP. The fill is four_regions which is world regions used to determine the color. The size is based on energy which is the energy usage. 

## Part 2: Import the q2data.csv dataset

```python
#Import the q2data.csv dataset.
df2 = pd.read_csv('C:/Users/alexa/OneDrive/Desktop/GSB544/data/q2data.csv')
```

```python
df2
```

## Part 3: Recreate the plot as best you can using plotnine

```python
#remove missing values
df2 = df2.dropna()
df2
```

```python
#understand why the error regarding non numeric values was occuring
df2['energy']
contains_k=df2['energy'].str.contains('k')
df2[contains_k]
```

```python
df2['energy'] = df2['energy'].astype(str)
energy_use = []

#replace those with 12.4k to 12400
for x in df2['energy']:
    if re.finditer('k', x):
        x = x.replace("k", "00")
        x = x.replace(".", "")
        energy_use.append(x)
    else:
        energy_use.append(x)  

df2['energy_use'] = energy_use

df2['energy_use'] = df2['energy_use'].astype(float)
```

```python
df2['energy_use']
```

```python
energy_values = [500,1100,1500, 1700,2500,3500,4500,8000,10000, 50000]
(
ggplot(df2,aes(x="exports", y="imports", fill = "four_regions"))
+scale_y_continuous(breaks = [50,100,150,200,250,300,350,400], limits = [0,450])
+scale_x_continuous(breaks = [20,40,60,80,100,120,140,160,180,200,220], limits = [0,240])
+geom_text (aes(label = '1997'), color = "lightgray", x=120,y=220, ha='center', va='center', size = 200)
+geom_point(aes(size='energy_use'))
+scale_size_area(max_size = 20, breaks =energy_values)
+theme_light()
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+labs(y = "Imports(\% of GDP)", x="Exports(\% of GDP)")
+labs(fill = "World Regions")
+labs(size = "Energy Use")    
+theme(figure_size = (10,6), axis_text_y=element_text(rotation=90))
)
```

References: https://www.w3schools.com/python/ref_string_replace.asp; https://stackoverflow.com/questions/47714540/plotnine-rotating-labels

## Part 4: What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why itâ€™s good or bad for visualizing these data.

```python
plot1 = (
ggplot(df2,aes(x="four_regions", y="imports", fill = "four_regions"))
+scale_y_continuous(breaks = [25,50,75,100,125, 150], limits = [0,150])
+geom_violin()
+theme_light()
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+labs(y = "Imports(\% of GDP)", x="World Region")
+labs(fill = "World Regions")
)

plot2= (
ggplot(df2,aes(x="four_regions", y="energy_use", fill = "four_regions"))
+scale_y_log10(breaks = [200,500,1000,2000,5000,10000,20000,50000], limits = [100,55000], labels = [200,500,100,2000,5000,'10k', '20k', '50k'])
+geom_violin()
+theme_light()
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+labs(y = "Energy Use", x="World Region")
+labs(fill = "World Regions")
)

plot3 =(
ggplot(df2,aes(x="four_regions", y="exports", fill = "four_regions"))
+geom_violin()
+theme_light()
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+labs(x="World Region", y = "Exports(\% of GDP)")
+labs(fill = "World Regions")
)

print(plot1)
print(plot2)
print(plot3)
```

Here there are 3 seperate graphs, this is similar to facetting however it is using a different y numeric variable for each. Plot 1 has x variable four_region and y variable imports and the fill is determined by four_region as well. Plot 2 has x variable four_region and y variable energy and fill is four_region. Lastly, plot 3 has x variable four_region, y variable exports, and fill four_region.

This is worse to visualize the data because you need to look at three seperate plots while the bubble plot is able to show it all on one graph. The three violin plats are helpful at breaking down each numeric variable based on world region but not helpful in understanding them all together. Violin plots can also be slightly difficult to interpret but helpful at showing medians and ranges of values.

## Part 4: What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why itâ€™s good or bad for visualizing these data.

```python
plot_1 = (
ggplot(df3,aes(x="four_regions", y="gdp", fill = "four_regions"))
+scale_y_log10(breaks = [200,500,1000,2000,5000,10000,20000,50000,100000],limits = [0,200000], labels = [200,500,1000,2000,5000,'10k','20k','50k', '100k'])
+geom_boxplot()
+theme_light()
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+labs( x="World Regions", y = "GDP/capita" )
+labs(fill = "World Regions")  
)

plot_2 = (
ggplot(df3,aes(x="four_regions", y="income", fill = "four_regions"))
+geom_boxplot()
+theme_light()
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+labs( x="World Regions", y = "Income" )
+labs(fill = "World Regions")  
)

plot_3 = (
ggplot(df3,aes(x="four_regions", y="internet_users", fill = "four_regions"))
+geom_boxplot()
+theme_light()
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+labs( x="World Regions", y = "Individuals using the Internet" )
+labs(fill = "World Regions")  
)

print(plot_1)
print(plot_2)
print(plot_3)
```

The first plot has aesthetics of x is four_region, y is gdp, and fill is four_region. Plot 2 has x of four_region, y of income, and fill is four_region. The last plot has x as four_region, y as internet_users, and fill as four_region.

The boxplots are helpful for understanding the median and quantiles of a numeric value against the categorical value. However, they require it to be on different plots so it is not as easy to understand how all four variables connect to each other. It is helpful to see the individual relationships between the numeric variable and world region. However, I don't think it is as helpful as seeing all of the variables on the same plot like the bubble plot created above.
