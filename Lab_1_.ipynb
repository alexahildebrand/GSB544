```python
#Recreate the plot as best you can using plotnine.

#income elevel 1, 2, 3, 4 on top
#per person gdp capita bottom
#do I need x scale to be 16k, 32k, 64k
#2010 went away
#add years to life expectancy
#dots are super small
#take out nan?
#missing rows issue?
#too many lines in background

color_dict = {'africa':'cyan',
             'americas':'lime',
             'europe':'yellow',
             'asia':'deeppink'}

population_values = [1000000,250000000,500000000,750000000,1000000000]

(
ggplot(df1,aes(x="income", y="life_exp", fill = "four_regions"))
+scale_y_continuous(breaks = np.arange(0,91,10), limits = [20,90])
+scale_x_log10(breaks =[500,1000,2000,4000,8000,16000,32000,64000])
+geom_text (aes(label = '2010'), color = "lightgray", x=8000,y=60, ha='center', va='center', size = 30)
#+geom_text (aes(label = 'per person'), color = "lightgray", x=60000,y=10,ha='center', va='center',size = 40)
+geom_point(aes(size='population'))
#+scale_size (range =(3,10))
+scale_size_area(breaks =population_values)
+theme_light()
#+theme(text = element_text(color="dimgray"),
     # axis_line_x =element_line(color = "darkgray"),
     # axis_line_y =element_line(color = "darkgray"),
      #line = element_line(color="lightgray", size =1.4))
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+labs(y = "Life Expectancy", x="Income")
+labs(fill = "World Regions")
+labs(size = "Population")    
)

```

#Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

The x variable is exports as a % of GDP and y variable is Imports as a % of GDP The fill is world regions which determines the color. The size is based on Energy use. 

```python
#Import the q2data.csv dataset.
df2 = pd.read_csv('C:/Users/alexa/OneDrive/Desktop/GSB544/data/q2data.csv')
```

```python
df2
```

```python
#recreate plot as best you can using plotnine
```

```python
# if there is a k, then regex to 1000
df2['energy'] = pd.to_numeric(df2['energy'])
```

```python
energy_values = [1100,1500, 1700,2500,3500,4500,8000,10000]
(
ggplot(df2,aes(x="exports", y="imports", fill = "four_regions"))
+scale_y_continuous(breaks = np.arange(0,401,50), limits = [50,400])
+scale_x_continuous(breaks =np.arange(0,221,20), limits = [20,220])
+geom_text (aes(label = '1997'), color = "lightgray", x=140,y=200, ha='center', va='center', size = 30)
+geom_point(aes(size='energy'))
#+scale_size (range =(3,10))
+scale_size_area(breaks =energy_values)
+theme_light()
#+theme(text = element_text(color="dimgray"),
     # axis_line_x =element_line(color = "darkgray"),
     # axis_line_y =element_line(color = "darkgray"),
      #line = element_line(color="lightgray", size =1.4))
+scale_color_manual(values = color_dict)
+scale_fill_manual(values = color_dict)
+labs(y = "Imports(\% of GDP)", x="Exports(\% of GDP)")
+labs(fill = "World Regions")
+labs(size = "Energy Use")    
)
```
